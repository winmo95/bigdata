```python
ppl_seoul = pd.read_excel('../data/population.xls', 
                          header = 2,
                          usecols = 'B,D,G,J,N',
                          encoding = "UTF-8")
```

```python
ppl_seoul.head()
```

```python
ppl_seoul.rename(columns ={ppl_seoul.columns[0]: '지역구',
                           ppl_seoul.columns[1]: '인구수',
                           ppl_seoul.columns[2]: '한국인',
                           ppl_seoul.columns[3]: '외국인',
                           ppl_seoul.columns[4]: '65세이상'}, inplace = True)
```

```python
ppl_seoul
```

```python
data_result = pd.merge(cctv_seoul, ppl_seoul, on = '지역구') #데이터 합병
```

```python
data_result.describe() #기술통계로 데이터 보기 (중요)
```

```python
del data_result['2013년도 이전']
del data_result['2014년']
del data_result['2015년']
del data_result['2016년']
```

```python
data_result.head()
```

```python
data_result.set_index('지역구',inplace = True)
```

```python
data_result.head()
```

```python
import numpy as np #통계를 쉽게 해주기 위해 사용하는 라이브러리 
# import 라이브러리 오류 발생시 => pip install package
np.corrcoef(data_result['인구수'], data_result['소계'])  # 상관계수 
#        인구수   소계
#인구수   1        0.236
# 소계   0.236      1              인구와 소계 사이에는 0.236정도의 상관계수가 있다.
```

```python
import matplotlib.pyplot as plt     #매트랩
data_result['소계'].plot(kind = 'bar' ,grid =True, figsize=(10,10))  #히스토그램 그리기
```
