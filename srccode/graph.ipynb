```python
%matplotlib inline
import matplotlib.pyplot as plt
import math
```

```python
Pi = 3.141592
xs,ys1,ys2 = [],[],[]
```

```python
for x in range(360) :
    xs.append(x)
    ys1.append(math.sin(Pi*x/180))
    ys2.append(math.cos(Pi*x/180))
```

```python
plt.plot(xs, ys1, label ='SIN')
plt.plot(xs, ys2, label ='COS')
plt.xlim([0,360])
plt.xlabel('X Value')
plt.ylabel('Y Value')
plt.title('Trigonometric Function')
plt.legend()
plt.show()
```
