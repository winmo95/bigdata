```python
import os
#os.getcwd() = 현재 주소 가져오기

from bs4 import BeautifulSoup
from urllib.request import urlopen
import re
import pandas as pd
```

```python
cafe50 = pd.read_csv('../data/best_sandwiches_in_chicago.csv', sep=',', encoding='UTF-8', index_col=0)
cafe50
```

```python
price = []
address = []

for i in cafe50.index[:]:
    html = urlopen(cafe50['More'][i])
    soup_tmp = BeautifulSoup(html,'html.parser')
    
    gettings = soup_tmp.find('p','addy').get_text()
    
    price.append(gettings.split()[0][:-1])
    address.append(''.join(gettings.split()[1:-2]))
    
    print(gettings)
```

```python
len(price), len(address)
```

```python
cafe50['Price'] = price
cafe50['Address'] = address

cafe50 = cafe50.loc[:,['Rank','Cafe','Menu','Price','Address']]

cafe50.head(5)
```

```python
cafe50.set_index('Rank', inplace=True)
```

```python
import folium
import pandas as pd
import googlemaps
from tqdm import tqdm_notebook
import numpy as np
```

```python
gmaps = googlemaps.Client(key=gmaps_key)
```

```python
lat = []
lng = []

for n in tqdm_notebook(cafe50.index):
    if cafe50['Address'][n] !='Multiple':
        target_name = cafe50['Address'][n]+', '+'Chicago'
        gmaps_output = gmaps.geocode(target_name)
        location_output = gmaps_output[0].get('geometry')
        lat.append(location_output['location']['lat'])
        lng.append(location_output['location']['lng'])
    
    else:
        lat.append(np.nan)
        lng.append(np.nan)
```

```python
len(lat), len(lng)
```

```python
cafe50['lat'] = lat
cafe50['lng'] = lng
cafe50.head()
```

```python
mapping = folium.Map(location =[cafe50['lat'].mean(), cafe50['lng'].mean()], zoom_start=11)

for n in cafe50.index:
    if cafe50['Address'][n] != 'Multiple':
        folium.Marker([cafe50['lat'][n], cafe50['lng'][n]], popup=cafe50['Cafe'][n]).add_to(mapping)
        
mapping
```
